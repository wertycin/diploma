```python
from scipy.fft import fft, ifft, fftfreq
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
START = 0
START_ = 0
END_ = 50
END = 50
N = 500
```

```python
x0 = np.linspace(START, START_, N, endpoint=False)
x1 = np.linspace(START_, END_, N, endpoint=False)
x2 = np.linspace(END_, END, N)
x = np.concatenate((x0, x1, x2))
y0 = np.zeros_like(x0)
y1 = np.sin(x1)
y2 = np.zeros_like(x2)
y = np.concatenate((y0, y1, y2))
y.size
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
fig.show()
```

```python
M = 1500
T = (x.max() - x.min()) / M
print(M, T)
y_ = fft(y)
x_ = fftfreq(M, T)[:M//2]
fig, ax = plt.subplots()
ax.set_xlim(xmin = 0, xmax = 5)
ax.plot(x_, 2.0 * np.pi / M * abs(y_[0:M//2]))
fig.show()
```

```python
from scipy.fft import fft, fftfreq
import numpy as np
# Number of sample points
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
xf = fftfreq(N, T)[:N//2]
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```
